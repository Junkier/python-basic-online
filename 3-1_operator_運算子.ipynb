```python
### 算術運算子

# 專門給 int / float / complex
a = 12
b = 25
print(a+b)
print(a-b)
print(a*b)
print(a/b)

# 指數 , ex: 2^10 --> 1024
print(2**10)
print(6**3)

# 整數除
print(108 / 100)
print(108 // 100)
print(2376 // 123)

# 取餘數 (module)
print(108 % 100)
print(2376 % 123)

print("-"*80)

# 專門給 str 
# + -> 字串合併
message5 = "嗨嗨" + "," + "今天下雨了"
print(message5)

# * -> 字串擴充
message6 = "-QQ-" *10
print(message6)
print("abc"*5)

print("-"*80)

#################################################
### 關於顯示 組合文字 出來
name1 = "Jeff"
name2 = "Leo"

# 單純使用 print
print(name1 , "和" , name2)

# 使用 .format 合併成大字串顯示
print( "{} 和 {}".format(name1, name2) )

# 使用 + 做字串合併
message7 = name1 + " 和 " + name2
print(message7)
```

```python
### 關係運算子 --- > ; < ; == ; !=
a = 3
b = 30
print(a>b)
print(a*3 >b)
print(a < b/2)

c = 90
print(a*3 == c)
print(b*3 == c)

name4 = "Jeff"
print(name4 == "Keven")  # 用問句的方式, 比較容易理解
print(name4 != "jeff") 

print("-"*80)

### 邏輯運算子 --- and ; or ; not
print(a>10 and b <60)
print(a % 2 == 0 or b % 2 == 0)  # a 是偶數 or b 是偶數

print("-"*80)

# 流程控制
if a>10:
    print("好, 今天我們下課！！！")
    
teacher = "Keven"

if teacher != "Jeff":
    print("今天不上課了喔嗚嗚嗚")
else:
    print("好, 我們開始上課。")
```
